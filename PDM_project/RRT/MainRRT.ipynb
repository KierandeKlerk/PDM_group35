```python
from Testmaps import map
from GridRRT_star import GridRRTstar
from tqdm import tqdm
import time
import numpy as np

```

```python
# Choose test map:
#   1: easy
#   2: normal
#   3: hard
#   4: using Grids
# Others are empty

start = np.array([5, 5])
goal = np.array([22, 50])
occupancy_grid = np.load("occupancy_grid.npy")
mapdim = (occupancy_grid.shape[0], occupancy_grid.shape[1])
dgoal = 3
dsearch = 5
dcheaper = 10
max_iter = 10000



#Computng map and path
graph = GridRRTstar(start, goal, mapdim, dgoal, dsearch, dcheaper, occupancy_grid, max_iter)
graph.makemap()
starttime = time.time()
pbar = tqdm(total = max_iter)

while graph.iterations():
    graph.expand()
    pbar.update(1)
pbar.close()
```
